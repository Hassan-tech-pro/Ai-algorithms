```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv(r'D:\Hassan\Projects\machine learning\linear regression 1\exercise\canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income/capita')
plt.scatter(df.year,df.income, color='red', marker='*')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["year"]],df.income)
```

```python
reg.predict([[2040],[5000]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel("year",fontsize=20)
plt.ylabel("income",fontsize=20)
plt.scatter(df.year,df.income,)
plt.plot(df.year,reg.predict(df[['year']]))
```
